```python
import torch
import torch.nn.functional as F
```

# dbnet

```python
from doctr.models import db_resnet50
model = db_resnet50(pretrained=True)
input_tensor = torch.rand((1, 3, 1024, 1024), dtype=torch.float32)
out = model(input_tensor)
```

```python
print(model)
```

```python
model=model.feat_extractor
```

```python
with torch.no_grad():
    out = model(input_tensor)
print(out)
```

```python
for i in range(4):
    print(out[f'{i}'].shape)
```

```python
deepest_feature= out['3']  # shape: [1, 2048, 32, 32]
pooled = F.adaptive_avg_pool2d(deepest_feature, (1, 1))  # shape: [1, 2048, 1, 1]

# Flatten to shape: [1, 2048]
feature_vector = pooled.view(pooled.size(0), -1)
print(feature_vector.shape)  # shape: [1, 2048]
```

# crnn

```python
from doctr.models import crnn_vgg16_bn
model = crnn_vgg16_bn(pretrained=True)
input_tensor = torch.rand(1, 3, 32, 128)
```

```python
print(model)
```

```python
model=model.feat_extractor
with torch.no_grad():
    out = model(input_tensor)
```

```python
print(out.shape)
```

```python
# x: [B, C, 1, W]
x=out
x = x.squeeze(2)           # [B, C, W]
x = x.permute(0, 2, 1)     # [B, W, C]
x = x.mean(dim=1)          # [B, C]  <-- global average over the sequence
print(x.shape)  # shape: [1, 512]
```

# doctr preprocessing

```python
from doctr.models import detection_predictor, recognition_predictor
from doctr.models.preprocessor import PreProcessor
```

```python
det_predictor = detection_predictor(
    arch="db_resnet50",
    pretrained=False,
    assume_straight_pages=True,
    symmetric_pad=True,
    preserve_aspect_ratio=True,
    batch_size=1,
)  # .cuda().half()  # Uncomment this line if you have a GPU
print(det_predictor.pre_processor)
#it is strange, the default preprocessor resize to a shape that is incompatible to the model (3,1024) instead of (1024,1024) ..
```

```python
transform=PreProcessor(
        (1024, 1024),
        batch_size=1,
        mean=(0.798, 0.785, 0.772),
        std=(0.264, 0.2749, 0.287),
        #preserve_aspect_ratio=True,
        #symmetric_pad=True,
    )
print(transform)
```

```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np

import numpy as np

# Create synthetic pages as NumPy arrays in HWC format
num_pages = 3
height, width, channels = 64, 64, 3

pages = [
    np.random.randint(0, 256, size=(height, width, channels), dtype=np.uint8)
    for _ in range(num_pages)
]

transformed_image=transform(pages)
```

```python
print(transformed_image[0].shape)  # shape: [3, 3, 64, 64]
```
