```python
import os
import pandas as pd
source_dir="D:\download\PD project\datasets\landlord\handwriting-recognition\\versions\\1"
val_dir=source_dir+"\\validation_v2\\validation"
train_dir=source_dir+'\\train_v2\\train'
test_dir='test_v2\\test'
```

```python
train_df = pd.read_csv(os.path.join(source_dir, "written_name_train_v2.csv"))
train_df['isTrain'] = 1
train_df.head(10)
```

```python
missing_values = train_df.isnull().sum()
print(missing_values)
```

```python
len(train_df)
```

```python
train_df = train_df.dropna()
train_df.reset_index(drop=True, inplace=True)
len(train_df)
```

```python
val_df = pd.read_csv(os.path.join(source_dir, "written_name_validation_v2.csv"))
val_df['isTrain'] = 0
val_df.head(10)
```

```python
missing_values = val_df.isnull().sum()
print(missing_values)
```

```python
print(len(val_df))
val_df = val_df.dropna()
val_df.reset_index(drop=True, inplace=True)
print(len(val_df))
```

```python
print(len(train_df)+len(val_df))
train_df = pd.concat([train_df, val_df], ignore_index=True)
train_df.reset_index(drop=True, inplace=True)
print(len(train_df))
train_df.head()
```

```python
train_df.loc[train_df['isTrain'] == 1, 'FILENAME'] = train_df.loc[train_df['isTrain'] == 1, 'FILENAME'].apply(lambda x: os.path.join(train_dir, x))
train_df.head()
```

```python
train_df.loc[train_df['isTrain'] == 0, 'FILENAME'] = train_df.loc[train_df['isTrain'] == 0, 'FILENAME'].apply(lambda x: os.path.join(val_dir, x))
train_df.head(-10)
```

```python
unique_identities_count = train_df['IDENTITY'].nunique()
print(unique_identities_count)
```

```python
print(len(train_df[train_df['IDENTITY']=='UNREADABLE']))
```

```python
top_identities = train_df['IDENTITY'].value_counts().head(200)
print(top_identities)
```

```python
train_df = train_df[~train_df['IDENTITY'].isin(['UNREADABLE', 'EMPTY'])]
train_df.reset_index(drop=True, inplace=True)
#are there other special keys??
```

```python
unique_identities = train_df['IDENTITY'].unique().tolist()
print(unique_identities)
print(len(unique_identities))
```

```python
import csv

output_file = os.path.join(source_dir, "unique_identities.csv")
with open(output_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["IDENTITY"])  # Add a header
    for identity in unique_identities:
        writer.writerow([identity])

print(f"File saved at: {output_file}")
```
